```python
#checking the number of missing values in the dataframe
air_quality_data.isnull().sum()
```

This shows that there are no missing values in the dataset. But the actaul missing values are replaced with '-200'

```python
#counting the number of times the value '-200' is repeated
air_quality_data.isin([-200]).sum(axis=0)
```

Firstly convert all the '-200' with NaN

Replace all the NaN values with the mean of that specific column

```python
air_quality_data = air_quality_data.replace(to_replace=-200, value=np.nan)
```

```python
air_quality_data.isnull().sum()
```

This shows the actual number of missing values

```python
air_quality_data.tail()
```

```python
# Fill missing values only in numeric columns with their respective means
air_quality_data = air_quality_data.fillna(air_quality_data.select_dtypes(include='number').mean())
```

```python
air_quality_data.tail()
```

```python
date_info = pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y')
print(date_info)
```

```python
time_info = time_info = pd.to_datetime(air_quality_data['Time'], format='%H.%M.%S').dt.time
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
#combining bith series to pandas dataframe
date_time =  pd.concat([date_info, time_info], axis=1)
```

```python
#combining date and time
date_time['ds'] = date_time['Date'].astype(str) + ' ' + date_time['Time'].astype(str)
```
